
Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset into pandas dataframe
credit_card_data = pd.read_csv('/content/Credit Card Fraud Detection data.csv')
```

```python
feature_names = credit_card_data.columns[1:-1]
column_data = credit_card_data[feature_names]
credit_card_dataframe = pd.DataFrame(column_data, columns=feature_names)
```

```python
# first five rows of the data
credit_card_data.head()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking for the missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution
credit_card_data['Class'].value_counts()
```

Here, 0 represents normal transaction and 1 represents fraud transaction

Data analysing

```python
# separating the data
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under sampling

```python
legit_sample = legit.sample(n=492)
```

```python
#concatenating two dataframe
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training by logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model evaluation

Accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy on test data: ', test_data_accuracy)
```
